# Zomato Data Analysis Project


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe= pd.read_csv('Zomato data .csv')
dataframe.head(10)
```

# Convert the datatype of rate column 

```python
def ChangeRate(value):
    value= str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(ChangeRate)
print(dataframe.head(10))
```

```python
dataframe.head(10)
```

```python
dataframe.info()
```

## Types of Restaurant 

```python
dataframe.head(10)
```

```python
sns.countplot(x=dataframe['listed_in(type)'],hue=dataframe['listed_in(type)'],palette='hot',legend=False )
plt.xlabel('Type of Restaurants')
```

#### Conclusion = Majority of the restaurant falls under Dinning Category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': grouped_data})
plt.plot(result,c='pink',marker='o')
plt.xlabel('Types of Restaurant',c='red',size=15)
plt.ylabel('Votes',c='red',size=15)
```

### Conclusion = Dinning Restaurants have received maximum votes.

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5,color='purple')
plt.title('Rating Distribution')
plt.show()
```

#### Conclusion= Majority Restaurants having ratings from 3.5 to 4.0

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data,hue=couple_data,palette='winter',legend=False)
plt.xlabel('average spending by couple')
```

#### Conclusion = Average Spending by couple is 300.

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe,hue='online_order',palette='bone',legend=False)
```

#### Conclusion = Offline orders have received low ratings than online orders

```python
dataframe.head()
```

```python
pivot_table= dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.xlabel('Online Orders',size=12)
plt.ylabel('Types of Restaurant',size=12)
plt.show()
```

#### Conclusion = Dining Restaurants primarily accepts offine orders,whereas Cafes primarily receive online orders.This suggests that clients preferred orfers in person at restaurants , but prefer online ordering at cafes.
